```python
#divide the feature variables and target variables for split the data
features=salary_data.drop('Salary',axis=1)
features
```

```python
#from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
encoder.fit_transform(salary_data.iloc[:,-1])
```

```python
salary_data['Salary']=encoder.fit_transform(salary_data.iloc[:,-1])
salary_data.head()
```

```python
target=salary_data[['Salary']]
target.value_counts()
```

## Spliting data

```python
#from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.7,random_state=100)
```

```python
# check the shape of training data and testing data
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# applying onehot encoding using get dummies
x_train=pd.get_dummies(x_train)
x_train
```

```python
x_test=pd.get_dummies(x_test)
x_test
```

```python
cols=x_train.columns
```

```python
#applying robust sacler
#from sklearn.preprocessing import RobustScaler,StandardScaler,MinMaxScaler,Normalizer
rbst_scale=RobustScaler()
x_train=rbst_scale.fit_transform(x_train)
x_test=rbst_scale.fit_transform(x_test)
```

```python
#put it array data to dataframe with columns name
x_train=pd.DataFrame(x_train,columns=[cols])
x_test=pd.DataFrame(x_test,columns=[cols])
x_train.head()
```

## Run SVM with default hyperparameters

```python
# SVM model
svm=SVC()
```

```python
# fit the model
svm.fit(x_train, y_train)

y_pred = svm.predict(x_test)

y_pred
```

```python
# check the accuracy score
accuracy_score(y_test,y_pred)
```

```python
y_test['pred_salary']=y_pred
y_test
```

```python
cm=(confusion_matrix(y_test['Salary'],y_pred))
cm
```

```python
# Print the Confusion Matrix and slice it into four pieces


cm = confusion_matrix(y_test['Salary'], y_pred)

print('Confusion matrix\n\n', cm)
print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
# visualize confusion matrix with seaborn heatmap

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])
plt.figure(figsize=(10,10))
sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
print(classification_report(y_test['Salary'],y_pred))
```

```python
def print_score(clf, x_train, y_train, x_test, y_test, train=True):
    if train:
        pred = clf.predict(x_train)
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{classification_report(y_train, pred)}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(x_test)
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test['Salary'], pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{classification_report(y_test['Salary'], pred)}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test['Salary'], pred)}\n")
```
