## Data Pre-Processing

```python
data = tree1.copy()
data.rename(columns={'Marital.Status':'Marital_Status', 'Taxable.Income':'Taxable_Income','Work.Experience':'Work_Experience','City.Population':'City_Population'}, inplace = True)
categorical_features = data.describe(include=["object",'category']).columns
categorical_features
```

```python
#from sklearn.preprocessing import LabelEncoder
#Creating dummy vairables of the categorical features
le = LabelEncoder()
for col in categorical_features:
        le.fit(data[col])
        data[col] = le.transform(data[col])
data.head()
```

```python
data.Taxable_Income.value_counts()
```

```python
features=data.drop('Taxable_Income',axis=1)
features
```

```python
target=data[['Taxable_Income']]
target.value_counts()
```

## Spliting the data 

```python
x_train,x_test,y_train,y_test= train_test_split(features,target,train_size=0.7,random_state=50)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# fit the model
dec_tree= DecisionTreeClassifier( criterion='gini', splitter='best',max_depth=8)
```

```python
dec_tree.fit(x_train,y_train)
```

```python
#prediction
y_pred= dec_tree.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

## Aplying Gridsearch for best parameters

```python
params= {'criterion':['gini','entropy'], 'splitter':['best','random'],'max_depth':[1,2,5,7,8,10,12,15]}
```

```python
grid_search= GridSearchCV(dec_tree,params)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
# heatmap for confusion matrix
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='0.0f')
```

```python
print(classification_report(y_test,y_pred))
```

```python
#visualize the decicion score 
#from sklearn.tree import plot_tree
plt.figure(figsize=(12,7))
plot_tree(dec_tree,max_depth=10, feature_names=features.columns,class_names=tree1['Taxable.Income'].values,label='all',filled=True,rounded=True);
```
